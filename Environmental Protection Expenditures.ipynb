```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/Users/smritilimbu/Downloads/Environmental_Protection_Expenditures.csv")
# Drop the specified column
data = data.drop(columns=["ObjectId", "CTS_Name"])
data
```

```python
#Identify the countries to keep

countries_to_keep = ["China, P.R.: Mainland", "Australia", "Switzerland", "Russian Federation",  "Japan", "Germany", "Canada", "United Kingdom", "Italy", "France"
]
# Filter the DataFrame to keep only the specified countries
data = data[data['Country'].isin(countries_to_keep)]
# Display the filtered DataFrame
data
```

```python
# Preprocessing Steps

# Forward fill and backfill to handle missing values
data_filled = data.fillna(method='ffill', axis=1).fillna(method='bfill', axis=1)

# Since we're going to compare expenditures across countries and over time,
# let's simplify the dataset by focusing on "Expenditure on environment protection" in Domestic Currency
# and then aggregate the data for visualization purposes.

# Filter data for "Expenditure on environment protection" and "Domestic Currency"
env_protection_data = data_filled[
    (data_filled['Indicator'] == "Expenditure on environment protection") &
    (data_filled['Unit'] == "Domestic Currency")
]
```

```python
# Convert 'Expenditure' to numeric after melting and before groupby operation
env_protection_data_melted = env_protection_data.drop(columns=['Indicator', 'Unit']).melt(id_vars='Country', var_name='Year', value_name='Expenditure')
env_protection_data_melted['Expenditure'] = pd.to_numeric(env_protection_data_melted['Expenditure'], errors='coerce')

# Recalculate the mean expenditure for each country across all available years for comparison
mean_expenditure_by_country_corrected = env_protection_data_melted.groupby('Country')['Expenditure'].mean().reset_index()


```

# Environmental Protection Expenditure by Country

```python
mean_expenditure_by_country_corrected
```

```python
# Visualization with data labels
plt.figure(figsize=(14, 7))
barplot = sns.barplot(data=mean_expenditure_by_country_corrected, x='Country', y='Expenditure')
plt.title('Environmental Protection Expenditure by Country')
plt.xlabel('Country')
plt.ylabel('Expenditure (Domestic Currency)')

# Add data labels
for p in barplot.patches:
    barplot.annotate(format(p.get_height(), '.2e'), 
                     (p.get_x() + p.get_width() / 2., p.get_height()), 
                     ha = 'center', va = 'center', 
                     xytext = (0, 10), 
                     textcoords = 'offset points')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

# Visualizations By Countries:

# United Kingdom

```python
# Select data for "United Kingdom" for trend analysis
uk_data = env_protection_data_melted[env_protection_data_melted['Country'] == 'United Kingdom']
uk_data['Year'] = uk_data['Year'].str[1:].astype(int)  # Convert year from string to integer
uk_data['Expenditure'] = uk_data['Expenditure'].astype(float)  # Ensure expenditure is float for plotting

# Plot the trend of environmental protection expenditure over time for "United Kingdom"
plt.figure(figsize=(14, 7))
sns.lineplot(data=uk_data.dropna(), x='Year', y='Expenditure')
plt.title('Environmental Protection Expenditure Over Time in United Kingdom')
plt.xlabel('Year')
plt.ylabel('Expenditure (Domestic Currency)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Calculate and visualize year-on-year percentage change for "United Kingdom"
uk_data['YoY Change'] = uk_data['Expenditure'].pct_change() * 100  # Calculate year-on-year percentage change

# Plot the year-on-year percentage change with data labels for "uk"
plt.figure(figsize=(14, 7))
barplot_uk = sns.barplot(data=uk_data.dropna(), x='Year', y='YoY Change')
plt.title('Year-on-Year Percentage Change in Environmental Protection Expenditure in uk')
plt.xlabel('Year')
plt.ylabel('Percentage Change')

# Add data labels
for p in barplot_uk.patches:
    height = p.get_height()
    barplot_uk.text(p.get_x() + p.get_width() / 2., height + 0.5, f'{height:.2f}%', ha="center")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
